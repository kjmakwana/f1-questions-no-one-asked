```python
import pandas as pd
```

```python
quali=pd.read_csv("data/qualifying.csv")
quali.head(10)
```

```python
#replace \N in q3 with 0:00.000
quali["q1"]=quali['q1'].replace('\\N', None)
quali["q2"]=quali['q2'].replace('\\N', None)
quali["q3"]=quali['q3'].replace('\\N', None)
quali.head(20)
```

```python
quali["best_time"]=quali["q3"].fillna(quali["q2"]).fillna(quali["q1"])
quali
```

```python
quali=quali.dropna(subset=["best_time"])
quali[quali["best_time"].isna()]
```

```python
quali["best_time_s"]=quali["best_time"].apply(lambda x: float(x.split(":")[0])*60+float(x.split(":")[1]))
```

```python
quali.groupby("raceId")["best_time_s"].quantile([0,.25,.50,.75,1]).unstack()
```

```python
monza_races=races[races["circuitId"]==14][["raceId","year"]]
```

```python
monza_races.describe()
```

```python
monza_quali=quali[quali["raceId"].isin(monza_races["raceId"])]
monza_quali
```

```python
monza_quali.raceId.nunique()
```

```python
monza_quali=monza_quali.merge(monza_races,on="raceId")
monza_quali
```

```python
monza_quali.describe()
```

```python
monza_quali["year"]=monza_quali["year"].astype(int)
```

```python
monza_modern_quali=monza_quali[monza_quali["year"]>2000]
```

```python
agg_quali=monza_modern_quali.groupby(["raceId","year"])["best_time_s"].agg(["min","max","mean","std","median"]).sort_index(level="year")
#boxplot this data and sort by year ascending using plotly
import plotly.express as px
#make this blot with a dark background
px.defaults.template="plotly_dark"
#make this a line plot with median and 95th quartiles on both sides as shaded areas


fig=px.line(agg_quali,x=agg_quali.index.get_level_values("year"),y="median",title="How fast (and close) have F1 cars become this millenium?",labels={"median":"Median Times (with Fieldspread)","year":"Year"},error_y="std")
fig.update_xaxes(title="Year",showgrid=True)
#show all values of year on x -axis 
fig.update_xaxes(tickmode="array",tickvals=agg_quali.index.get_level_values("year").unique())
# fig=px.box(monza_modern_quali,x="year",y="best_time_s",title="Monza Qualifying Times",labels={"best_time_s":"Best Time (s)","year":"Year"})
fig.show()

```
