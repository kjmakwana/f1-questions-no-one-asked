```python
import pandas as pd
import numpy as np  
import plotly.express as px
```

```python
drivers = pd.read_csv('data/drivers.csv')
drivers.head()
```

```python
drivers.info()
```

```python
results=pd.read_csv("data/results.csv")
results.head()
```

```python
results.describe()
```

```python
results.info()
```

```python
wins=results[results["position"]=="1"]
wins.describe()
```

```python
driver_wins=drivers.merge(wins, on="driverId")
```

```python
driver_wins=driver_wins.groupby("driverRef")["raceId"].count().sort_values(ascending=False)
```

```python
drivers=drivers.merge(driver_wins, on="driverRef", how="left")
```

```python
drivers.rename(columns={"raceId":"wins"}, inplace=True)
drivers["wins"].fillna(0, inplace=True)
```

```python
drivers
```

```python
nation_wins=drivers.groupby("nationality")["wins"].sum().sort_values(ascending=False)
```

```python
drivers[(drivers["nationality"]=="Australian")]
```

```python
nation_wins
```

```python
drivers_wo_win=drivers[drivers["wins"]==0]
drivers_wo_win
```

```python
nations_wo_wins=drivers_wo_win.groupby("nationality")["driverId"].count().sort_values(ascending=False)
nations_wo_wins
```

```python
nation_wins_nil=nation_wins[nation_wins==0].index
nation_wins_nil
```

```python
nations_with_most_drivers_wo_wins=nations_wo_wins.loc[nation_wins_nil].to_frame().reset_index().sort_values(by="driverId", ascending=False)
# nations_with_most_drivers_wo_wins.setcolumns(["Nationality","Drivers"])
nations_with_most_drivers_wo_wins.columns=["Nationality","Drivers"]
nations_with_most_drivers_wo_wins
```

```python
nations_with_most_drivers_wo_wins.to_excel("ready_to_plot/nations_with_most_drivers_wo_wins.xlsx", index=False)
```
