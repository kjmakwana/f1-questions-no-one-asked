```python
from urllib.request import urlopen
import json
import pandas as pd
import plotly.express as px
```

```python
url="https://api.openf1.org/v1"
table="meetings"
```

```python
monza23 = json.loads(urlopen(url+"/"+table+"?year=2023&location=Monza").read().decode('utf-8'))
monza23
```

```python
max23_laps=json.loads(urlopen(url+"/"+"laps?meeting_key=1218&driver_number=1&session_key=9157").read().decode('utf-8'))
max23_laps=pd.DataFrame(max23_laps)
max23_laps.head()
```

```python
max24_laps=json.loads(urlopen(url+"/"+"laps?meeting_key=1244&driver_number=1&session_key=9590").read().decode('utf-8'))
max24_laps=pd.DataFrame(max24_laps)
max24_laps.head()
```

```python
max_monza_laps=max23_laps[["lap_duration","lap_number"]].merge(max24_laps[["lap_duration","lap_number"]], on="lap_number", suffixes=('_23', '_24'))
max_monza_laps.head()
```

```python
max23_stints=json.loads(urlopen(url+"/"+"stints?meeting_key=1218&driver_number=1&session_key=9157").read().decode('utf-8'))
max23_stints=pd.DataFrame(max23_stints)
max23_stints.head()
```

```python
max24_stints=json.loads(urlopen(url+"/"+"stints?meeting_key=1244&driver_number=1&session_key=9590").read().decode('utf-8'))
max24_stints=pd.DataFrame(max24_stints)
max24_stints.head()
```

```python
max23_stints["lap_number"]=max23_stints.apply(lambda x: list(range(x["lap_start"],x["lap_end"]+1)), axis=1)
max23_stints=max23_stints.explode("lap_number")
max23_stints
```

```python
# make the max24_stints table with lap number from 1 to 54 from lap_start to lap_end and have the corresponding compound
max24_stints["lap_number"]=max24_stints.apply(lambda x: list(range(x["lap_start"],x["lap_end"]+1)), axis=1)
max24_stints=max24_stints.explode("lap_number")
max24_stints
```

```python
max_stints=max23_stints[["lap_number","compound"]].merge(max24_stints[["lap_number","compound"]], on="lap_number", suffixes=('_23', '_24'),how="outer")
max_stints.head()
```

```python
max_monza_laps=max_monza_laps.merge(max_stints, on="lap_number")
```

```python
max_monza_laps.head()
```

```python
#update the plot to have a black background and have a different color line for tyre compounds, White for HARD and yellow for MEDIUM
fig=px.line(max_monza_laps, x="lap_number", y=["lap_duration_23", "lap_duration_24"], title="Max Verstappen Monza 2023 vs 2024", template="plotly_dark", color="compound_23")
# fig=px.line(max_monza_laps, x="lap_number", y=["lap_duration_23", "lap_duration_24"], title="Max Verstappen Monza 2023 vs 2024")
#the lines should be of different hues for different years but stints should be of the same color
fig=px.line(max_monza_laps, x="lap_number", y=["lap_duration_23", "lap_duration_24"], title="Max Verstappen Monza 2023 vs 2024", color="compound_23")
#colors are not white and yellow
fig=px.line(max_monza_laps, x="lap_number", y=["lap_duration_23", "lap_duration_24"], title="Max Verstappen Monza 2023 vs 2024", template="plotly_dark", color_discrete_map={"HARD":"white","MEDIUM":"yellow"})
fig.update_coloraxes(colorbar_title="Tyre compound")
fig.update_traces(mode="markers+lines")
fig.update_yaxes(title_text="Lap duration (s)")
fig.update_xaxes(title_text="Lap number")
fig.show()
```
